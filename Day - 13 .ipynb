## 6) filter for different condition,come up with insights from the dataset,basic and distribution plots

```python
creditcard['parsed_time'] = pd.to_datetime(creditcard['Time'])
```

```python
creditcard['parsed_time'][:-10:-1]
```

```python
fraud = creditcard[creditcard['Class'] == 1]
```

```python
fraud
```

```python
normal = creditcard[creditcard['Class']  == 0]
```

```python
normal
```

```python
print(normal.shape,fraud.shape)
```

```python
fig = plt.figure()
```

```python
fig,(ax1,ax2) = plt.subplots(2,sharex=True) # share of x-axis values for two subplots as same
fig.suptitle('Amount per transactions by class') # suptitle to create title for image
bins= 100   # intervals width of data = bins, suitable no. is square root of total values in data we take on plotting
ax1.hist(fraud.Amount, bins = bins)
ax1.set_title('fraud')
ax2.hist(normal.Amount,bins = bins)
ax2.set_title('normal')
plt.xlabel('Amount($)')
plt.ylabel('number of transactions')
plt.xlim((0,10000))
plt.yscale('log')
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(2,sharex=True) 
fig.suptitle('Time of transaction vs Amount by class')
ax1.scatter(fraud.Time,fraud.Amount )
ax1.set_title('fraud')
ax2.scatter(normal.Time,normal.Amount)
ax2.set_title('normal')
plt.xlabel('time in seconds')
plt.ylabel('Amount($)')
plt.show()
```
