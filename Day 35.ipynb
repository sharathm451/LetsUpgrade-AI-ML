# Use the user dataset perform random forest and compare with decision tree

```python
import numpy as np 
import pandas as pd 
```

```python
dataset = pd.read_csv(r"E:\AI-ML PYTHON\User_Data - Arindam Dev.csv")
```

```python
dataset.head()
```

```python
X = dataset.values[:,[2,3]] 
y = dataset.values[:,4]
```

```python
X
```

```python
y
```

```python
# datatype(dtype) of y is object so convert to int
y=y.astype('int')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 42,test_size = 0.25)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train) 
```

```python
X_test= sc.transform(X_test)  
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier( max_depth=2, 
                             random_state = 0)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import tree
```

```python
tree.plot_tree(clf)
```

```python
from sklearn.tree import export_text
text_representation = tree.export_text(clf)
print(text_representation)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
print(f"accuracy{accuracy:6.2f}")
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test,y_pred)
print(report)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 50)
classifier.fit(X_train,y_train) 
```

```python
y_pred = classifier.predict(X_test)
```

```python
print("confusion matrix \n")
matrix= confusion_matrix(y_test,y_pred)
print(cm)
print("classification report \n")
reports = classification_report(y_test,y_pred)
print(reports)
```

```python
accuracy = accuracy_score(y_test,y_pred)
print(f"accuracy{accuracy:6.2f}")
```
